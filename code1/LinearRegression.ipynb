```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
train_data = pd.read_csv("train.csv")
test_data = pd.read_csv("test.csv")
```

```python
train_data.head(),test_data.head()
```

```python
train_data.shape, test_data.shape
```

```python
train_data.isna().sum()
```

```python
test_data.isna().sum()
```

```python
train_data.dropna(inplace = True)
```

```python
train_data.describe()
```

```python
plt.title("Scatter Plot of Train Data")
plt.scatter(train_data['x'],train_data['y'])
plt.show()
```

```python
plt.title("Scatter plot of test data")
plt.scatter(test_data['x'],test_data['y'])
plt.show()
```

```python
x_train = train_data.iloc[:,0].values.reshape(-1,1)
y_train = train_data.iloc[:,1].values.reshape(-1,1)

x_test = test_data.iloc[:,0].values.reshape(-1,1)
y_test = test_data.iloc[:,1].values.reshape(-1,1)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
```

```python
plt.title("Prediction on test data")
plt.plot(x_test, prediction, label='Linear Regression', color='r')
plt.scatter(x_test, y_test, label='A¡¡ctual Test Data')
plt.legend()
plt.show()
```

```python
accuracy = model.score(y_test,prediction)
accuracy
```
